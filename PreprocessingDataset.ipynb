Splitting into RCC RMLO LCC LMLO

```python
import csv
import cv2

with open('/content/drive/MyDrive/train.csv', 'r') as f:
    d_reader = csv.DictReader(f, delimiter=';')
    headers = d_reader.fieldnames
    print(headers)
    iterator = 0
    greseli = []
    for line in d_reader:
      if iterator>=0:
        try:
          print(line['patient_id'])
          print(line['image_id'])
          image = cv2.imread('/content/Dataset_Unprocessed/'+line['patient_id']+'_'+line['image_id']+'.png')
          gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
          thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU + cv2.THRESH_BINARY)[1]
          cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
          cnts = cnts[0] if len(cnts) == 2 else cnts[1]
          cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
          for c in cnts:
            x,y,w,h = cv2.boundingRect(c)
            ROI = image[y:y+h, x:x+w]
            break
          image = ROI
          if line['laterality']=="L" and line['view']=="CC" and line['cancer']=="0":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/LCC/NEGATIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("LCC False")
          if line['laterality']=="L" and line['view']=="CC" and line['cancer']=="1":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/LCC/POSITIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("LCC True")
          if line['laterality']=="R" and line['view']=="CC" and line['cancer']=="0":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/RCC/NEGATIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("RCC False")
          if line['laterality']=="R" and line['view']=="CC" and line['cancer']=="1":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/RCC/POSITIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("RCC True")
          if line['laterality']=="L" and line['view']=="MLO" and line['cancer']=="0":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/LMLO/NEGATIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("LMLO False")
          if line['laterality']=="L" and line['view']=="MLO" and line['cancer']=="1":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/LMLO/POSITIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("LMLO True")
          if line['laterality']=="R" and line['view']=="MLO" and line['cancer']=="0":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/RMLO/NEGATIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("RMLO False")
          if line['laterality']=="R" and line['view']=="MLO" and line['cancer']=="1":
            cv2.imwrite('/content/Licenta/Processed_RSNA2/RMLO/POSITIVE/'+line['patient_id']+line['image_id']+'.png',ROI)
            print("RMLO True")
        except:
          greseli.append(line['patient_id']+line['image_id'])
          print("eroare")
      iterator=iterator+1
    print(greseli)
```
