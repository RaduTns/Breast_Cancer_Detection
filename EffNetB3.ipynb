```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split

data_dir = '/content/content/Licenta/Processed_RSNA2/LCC'

# Get list of all image file paths
image_paths = []
for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith('.png'):
            image_paths.append(os.path.join(root, file))

# Generate labels from file paths
labels = [os.path.basename(os.path.dirname(path)) for path in image_paths]

# Split data into training and test sets
train_paths, test_paths, train_labels, test_labels = train_test_split(image_paths, labels, test_size=0.2) #, random_state=40

# Split training data into training and validation sets
train_paths, val_paths, train_labels, val_labels = train_test_split(train_paths, train_labels, test_size=0.2) #, random_state=40

# Create a DataFrame to store the file paths and labels
train_df = pd.DataFrame({
    'path': train_paths,
    'label': train_labels
})
val_df = pd.DataFrame({
    'path': val_paths,
    'label': val_labels
})
test_df = pd.DataFrame({
    'path': test_paths,
    'label': test_labels
})

# Print number of images in each set
print(f'Training set: {len(train_df)} images')
print(f'Validation set: {len(val_df)} images')
print(f'Test set: {len(test_df)} images')
```

```python
print('Training set:')
print(train_df['label'].value_counts())

print('\nValidation set:')
print(val_df['label'].value_counts())

print('\nTest set:')
print(test_df['label'].value_counts())
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest',

)
test_datagen = ImageDataGenerator(rescale=1./255)

# Define the batch size
batch_size = 8

# Create data generators
train_generator = train_datagen.flow_from_dataframe(
    train_df,
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    color_mode='grayscale',
    batch_size=batch_size,
    class_mode='binary'
)
val_generator = test_datagen.flow_from_dataframe(
    val_df,
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    color_mode='grayscale',
    batch_size=batch_size,
    class_mode='binary'
)
test_generator = test_datagen.flow_from_dataframe(
    test_df,
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    color_mode='grayscale',
    batch_size=batch_size,
    class_mode='binary'
)

# Select a subset of the negative cases to reduce their numbers
neg_df = train_df[train_df['label'] == 'NEGATIVE']
pos_df = train_df[train_df['label'] == 'POSITIVE']

print("pos_df: ", len(pos_df))

# Generate 4 augmented versions for each positive image
augmentation_factor = 4

neg_sample_size = len(pos_df) * (augmentation_factor+1)  # Select augmentation_factor+1 as many negative cases as positive cases
neg_df_sampled = neg_df.sample(n=neg_sample_size, random_state=42)
print("neg_df_sampled: ", len(neg_df_sampled))

# Concatenate the positive and sampled negative dataframes
train_df_augmented = pd.concat([pos_df, neg_df_sampled], axis=0)




# Apply augmentation to positive cases only
pos_df_augmented = pd.DataFrame(columns=['path', 'label'])
for i, row in pos_df.iterrows():
    img_path = row['path']
    label = row['label']
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224), color_mode='grayscale')
    x = tf.keras.preprocessing.image.img_to_array(img)
    x = x.reshape((1,) + x.shape)
    for j, batch in enumerate(train_datagen.flow(x, batch_size=1)):
        img_aug = tf.keras.preprocessing.image.array_to_img(batch[0])
        img_aug_path = f'{img_path.split(".")[0]}_aug{i}_{j}.jpeg'
        img_aug.save(img_aug_path)
        pos_df_augmented = pos_df_augmented.append({'path': img_aug_path, 'label': label}, ignore_index=True)
        if j + 1 >= augmentation_factor:
            break

# Concatenate the augmented positive dataframe with the augmented negative dataframe
train_df_augmented = pd.concat([train_df_augmented, pos_df_augmented], axis=0)

# Shuffle the augmented dataframe
train_df_augmented = train_df_augmented.sample(frac=1, random_state=42)

# Print number of images in each set after augmentation
print(f'Training set: {len(train_df_augmented)} images')
print(f'Validation set: {len(val_df)} images')
print(f'Test set: {len(test_df)} images')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import applications
from tensorflow.keras.models import Model


def create_model():
    model = applications.EfficientNetB3(include_top=False, weights=None, input_shape=(224,224,1))
    x = model.output
    x = Flatten()(x)
    x = Dropout(0.5)(x)
    x = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=model.input, outputs=x)

    return model
```

```python
from tensorflow.keras.metrics import BinaryAccuracy

def true_positives(y_true, y_pred):
    # Calculate the true positives
    tp = tf.reduce_sum(tf.cast(tf.logical_and(tf.equal(y_true, 1), tf.equal(tf.round(y_pred), 1)), dtype=tf.float32))
    return tp

def true_negatives(y_true, y_pred):
    # Calculate the true negatives
    tn = tf.reduce_sum(tf.cast(tf.logical_and(tf.equal(y_true, 0), tf.equal(tf.round(y_pred), 0)), dtype=tf.float32))
    return tn
```

```python
from tensorflow.keras.optimizers.legacy import Adam

log_dir = '/content/drive/MyDrive/Licenta/Logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)

train_generator_augmented = train_datagen.flow_from_dataframe(
    train_df_augmented,
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    color_mode='grayscale',
    batch_size=batch_size,
    class_mode='binary'
)

class_weights = {0: 1., 1: 2.}

optimizer = Adam(learning_rate=0.0005)
model = create_model()
model.summary()
model.compile(optimizer=optimizer,
            loss='binary_crossentropy',
            metrics=['accuracy', true_positives, true_negatives, BinaryAccuracy(threshold=0.25)])
hist = model.fit(train_generator_augmented,epochs=20,validation_data=val_generator, steps_per_epoch=len(train_generator_augmented), validation_steps=len(val_generator))
model.save('./ResNet50LCC_model.h5', overwrite=True)
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy, TruePositives, TrueNegatives, FalsePositives, FalseNegatives
from tqdm import tqdm

pre = Precision()
re = Recall()
acc = BinaryAccuracy()
tp = TruePositives()
tn = TrueNegatives()
fp = FalsePositives()
fn = FalseNegatives()

for i, batch in tqdm(enumerate(test_generator), total=len(test_generator)):
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
    tp.update_state(y, yhat)
    tn.update_state(y, yhat)
    fp.update_state(y, yhat)
    fn.update_state(y, yhat)
    if(i>len(test_generator)):
     break

print("Precision:", pre.result().numpy())
print("Recall:", re.result().numpy())
print("Binary Accuracy:", acc.result().numpy())
print("True Positives:", tp.result().numpy())
print("True Negatives:", tn.result().numpy())
print("False Positives:", fp.result().numpy())
print("False Negatives:", fn.result().numpy())
```
