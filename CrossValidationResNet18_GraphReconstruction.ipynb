```python
import os
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from sklearn.model_selection import train_test_split, KFold
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import StratifiedKFold

# Same as before
data_dir = '/content/content/Licenta/Processed_RSNA2'
image_paths = []
for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith('.png'):
            image_paths.append(os.path.join(root, file))
labels = [os.path.basename(os.path.dirname(path)) for path in image_paths]
train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(image_paths, labels, test_size=0.1, random_state=90, stratify=labels)

print(f'Training set: {len(train_val_paths)} images')
print(f'Test set: {len(test_paths)} images')

positive_dir = '/content/POSITIVE'
positive_paths = []
for root, dirs, files in os.walk(positive_dir):
    for file in files:
        if file.endswith('.png'):
            positive_paths.append(os.path.join(root, file))
positive_labels = [os.path.basename(os.path.dirname(path)) for path in positive_paths]
positive_df = pd.DataFrame({'path': positive_paths, 'label': positive_labels})
train_val_df = pd.concat([pd.DataFrame({'path': train_val_paths, 'label': train_val_labels}), positive_df], ignore_index=True)

# Custom dataset
class CustomDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_name = self.dataframe.iloc[idx, 0]
        image = Image.open(img_name).convert('L')
        label = 0 if self.dataframe.iloc[idx, 1] == 'NEGATIVE' else 1

        if self.transform:
            image = self.transform(image)

        return image, label

# Data transformations
transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
])

test_transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
])
test_labels_count = pd.Series(test_labels).value_counts()
print(f"Test Set: NEGATIVE: {test_labels_count.get('NEGATIVE', 0)}, POSITIVE: {test_labels_count.get('POSITIVE', 0)}")
# Stratified KFold cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)

# Extract labels for stratification
X = train_val_df['path'].values
y = train_val_df['label'].apply(lambda x: 0 if x == 'NEGATIVE' else 1).values

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    print(f'Fold {fold + 1}')

    train_fold_df = train_val_df.iloc[train_idx].copy()
    val_fold_df = train_val_df.iloc[val_idx].copy()

    # Count and print number of positive and negative cases in each split
    train_counts = train_fold_df['label'].value_counts()
    val_counts = val_fold_df['label'].value_counts()
    print(f"Training Split: NEGATIVE: {train_counts.get('NEGATIVE', 0)}, POSITIVE: {train_counts.get('POSITIVE', 0)}")
    print(f"Validation Split: NEGATIVE: {val_counts.get('NEGATIVE', 0)}, POSITIVE: {val_counts.get('POSITIVE', 0)}")

    train_dataset = CustomDataset(dataframe=train_fold_df, transform=transform)
    val_dataset = CustomDataset(dataframe=val_fold_df, transform=test_transform)

    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=0)

   # Model
    model = models.resnet18(pretrained=False)  # Set pretrained=False because you are changing the input dimensions

    # Modify the input layer to accept single-channel (grayscale) images
    model.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
    # Check if CUDA is available
    if torch.cuda.is_available():
        print("CUDA is available. PyTorch is using GPU.")
    else:
        print("CUDA is not available. PyTorch is using CPU.")
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 1)  # Binary classification
    model = model.to('cuda')  # Move model to GPU

    # Loss and optimizer
    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.0001)

    # Training the model
    num_epochs = 3
    patience = 7  # Patience for early stopping
    best_loss = None
    epochs_no_improve = 0
    training_losses = []
    validation_losses = []
    precisions_train = []
    recalls_train = []
    precisions_val = []
    recalls_val = []

    for epoch in range(num_epochs):
        # Training
        model.train()
        total_training_loss = 0
        all_training_labels = []
        all_training_preds = []
        for inputs, labels in tqdm(train_loader, desc=f'Epoch {epoch + 1} Training'):
            inputs, labels = inputs.to('cuda'), labels.to('cuda').float().unsqueeze(1)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            total_training_loss += loss.item()
            training_preds = (torch.sigmoid(outputs).detach().cpu().numpy() > 0.5).astype(int)
            all_training_preds.extend(training_preds)
            all_training_labels.extend(labels.cpu().numpy())
        average_training_loss = total_training_loss / len(train_loader)
        training_losses.append(average_training_loss)

        # Calculate precision and recall for training set
        training_precision = precision_score(all_training_labels, all_training_preds)
        training_recall = recall_score(all_training_labels, all_training_preds)
        precisions_train.append(training_precision)
        recalls_train.append(training_recall)

        # Validation
        model.eval()
        total_validation_loss = 0
        all_validation_labels = []
        all_validation_preds = []
        with torch.no_grad():
            val_losses = []
            for inputs, labels in tqdm(val_loader, desc=f'Epoch {epoch + 1} Validation'):
                inputs, labels = inputs.to('cuda'), labels.to('cuda').float().unsqueeze(1)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_losses.append(loss.item())
                validation_preds = torch.sigmoid(outputs).cpu().numpy() > 0.5
                all_validation_preds.extend(validation_preds)
                all_validation_labels.extend(labels.cpu().numpy())

        avg_val_loss = sum(val_losses) / len(val_losses)
        validation_losses.append(avg_val_loss)

        # Calculate precision and recall for validation set
        validation_precision = precision_score(all_validation_labels, all_validation_preds)
        validation_recall = recall_score(all_validation_labels, all_validation_preds)
        precisions_val.append(validation_precision)
        recalls_val.append(validation_recall)

        print(f'Epoch [{epoch+1}/{num_epochs}], '
              f'Training Loss: {average_training_loss:.4f}, Training Precision: {training_precision:.4f}, '
              f'Training Recall: {training_recall:.4f}, '
              f'Validation Loss: {avg_val_loss:.4f}, Validation Precision: {validation_precision:.4f}, '
              f'Validation Recall: {validation_recall:.4f}')

        # Checkpoint and Early Stopping
        if best_loss is None:
            best_loss = avg_val_loss
            torch.save(model.state_dict(), f'model_fold_{fold + 1}_best.pth')
        elif avg_val_loss < best_loss:
            epochs_no_improve = 0
            best_loss = avg_val_loss
            torch.save(model.state_dict(), f'model_fold_{fold + 1}_best.pth')
            print(f'Validation loss decreased. Saving the model to model_fold_{fold + 1}_best.pth')
        else:
            epochs_no_improve += 1
            if epochs_no_improve == patience:
                print('Early stopping')
                break
    actual_epochs = len(training_losses)
    # Plot Training Loss
    plt.figure(figsize=(6, 4))
    plt.plot(range(1, actual_epochs + 1), training_losses, label='Training Loss')
    plt.plot(range(1, actual_epochs + 1), validation_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Training and Validation Loss')
    plt.show()

    # Plot Training Precision
    plt.figure(figsize=(6, 4))
    plt.plot(range(1, actual_epochs + 1), precisions_train, label='Training Precision')
    plt.plot(range(1, actual_epochs + 1), precisions_val, label='Validation Precision')
    plt.xlabel('Epochs')
    plt.ylabel('Precision')
    plt.legend()
    plt.title('Training and Validation Precision')
    plt.show()

    # Plot Training Recall
    plt.figure(figsize=(6, 4))
    plt.plot(range(1, actual_epochs + 1), recalls_train, label='Training Recall')
    plt.plot(range(1, actual_epochs + 1), recalls_val, label='Validation Recall')
    plt.xlabel('Epochs')
    plt.ylabel('Recall')
    plt.legend()
    plt.title('Training and Validation Recall')
    plt.show()
```

```python
from sklearn.metrics import confusion_matrix

# Testing the model on test dataset
test_dataset = CustomDataset(dataframe=pd.DataFrame({'path': test_paths, 'label': test_labels}), transform=test_transform)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=0)

for fold in range(5):
    # Load the saved model
    model = models.resnet18(pretrained=False)
    model.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 1)
    model = model.to('cuda')
    model.load_state_dict(torch.load(f'model_fold_{fold + 1}_best.pth'))

    # Evaluate on test dataset
    model.eval()
    all_test_labels = []
    all_test_preds = []
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc=f'Testing fold {fold + 1}'):
            inputs, labels = inputs.to('cuda'), labels.to('cuda').float().unsqueeze(1)
            outputs = model(inputs)
            test_preds = torch.sigmoid(outputs).cpu().numpy() > 0.03
            all_test_preds.extend(test_preds)
            all_test_labels.extend(labels.cpu().numpy())

    # Calculate the confusion matrix
    tn, fp, fn, tp = confusion_matrix(all_test_labels, all_test_preds).ravel()

    # Calculate precision, accuracy, and recall
    precision = precision_score(all_test_labels, all_test_preds)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    recall = recall_score(all_test_labels, all_test_preds)

    print(f'Metrics for fold {fold + 1}:')
    print(f'True Positives: {tp}, True Negatives: {tn}, False Positives: {fp}, False Negatives: {fn}')
    print(f'Precision: {precision:.4f}, Accuracy: {accuracy:.4f}, Recall: {recall:.4f}\n')
```

```python
from collections import Counter

# Testing the model on test dataset
test_dataset = CustomDataset(dataframe=pd.DataFrame({'path': test_paths, 'label': test_labels}), transform=test_transform)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=0)

# Store predictions from all models
all_models_test_preds = []

for fold in range(5):
    # Load the saved model
    model = models.resnet18(pretrained=False)
    model.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 1)
    model = model.to('cuda')
    model.load_state_dict(torch.load(f'model_fold_{fold + 1}_best.pth'))

    # Evaluate on test dataset
    model.eval()
    all_test_labels = []
    model_test_preds = []
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc=f'Testing fold {fold + 1}'):
            inputs, labels = inputs.to('cuda'), labels.to('cuda').float().unsqueeze(1)
            outputs = model(inputs)
            test_preds = torch.sigmoid(outputs).cpu().numpy() > 0.03
            model_test_preds.extend(test_preds)
            all_test_labels.extend(labels.cpu().numpy())
    all_models_test_preds.append(model_test_preds)

all_models_test_preds = list(zip(*all_models_test_preds))

# Perform majority voting
majority_voted_preds = []
for pred_set in all_models_test_preds:
    # Convert boolean numpy arrays within the tuple to int (True -> 1, False -> 0)
    pred_set_list = [int(p) for sublist in pred_set for p in sublist]
    majority_vote = Counter(pred_set_list).most_common(1)[0][0]
    majority_voted_preds.append(majority_vote)

# Calculate the confusion matrix
tn, fp, fn, tp = confusion_matrix(all_test_labels, majority_voted_preds).ravel()

# Calculate precision, accuracy, and recall
precision = precision_score(all_test_labels, majority_voted_preds)
accuracy = (tp + tn) / (tp + tn + fp + fn)
recall = recall_score(all_test_labels, majority_voted_preds)

print('Metrics for ensemble with majority voting:')
print(f'True Positives: {tp}, True Negatives: {tn}, False Positives: {fp}, False Negatives: {fn}')
print(f'Precision: {precision:.4f}, Accuracy: {accuracy:.4f}, Recall: {recall:.4f}\n')
```
